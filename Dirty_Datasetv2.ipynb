```python
import numpy as np
import pandas as pd
import os
import re
import missingno as msn
```

```python
df_src= pd.read_csv("Dirty_Dataset.csv")
df_src.head()
```

```python
df_src.info()
```

```python
df = df_src.copy()
pattern = "(\[[0-9]*[a-z]*\])"
df.Peak = df.Peak.apply(lambda x: re.sub(pattern, "",  str(x)))
df["All Time Peak"] = df["All Time Peak"].apply(lambda x: re.sub(pattern, "",  str(x)))
df['Tour title'] = df['Tour title'].apply(lambda x: re.sub(pattern, "",  str(x)))
df['Actual\xa0gross'] = df["Actual\xa0gross"].apply(lambda x: re.sub(pattern, "",str(x)))


```

```python
df['Year_start'] = df['Year(s)'].apply(lambda x: x[0:4])
df['Year_end'] = df['Year(s)'].apply(lambda x: x[-4:])
df.head()
```

```python
DROPLIST = ['Year(s)', 'Ref.']

df.drop(DROPLIST, axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
cols_obj = df.select_dtypes(['object']).columns
cols_obj
```

```python
for cols  in cols_obj:
    df[cols] = df[cols].apply(lambda x: x.replace("nan", "0"))
df.head()
```

```python
for cols in [cols_obj[2], cols_obj[3], cols_obj[6], cols_obj[7], cols_obj[8]]:
    df[cols] = df[cols].apply(lambda x: str(x).replace('$', "").replace(",", ""))
    df[cols] = df[cols].astype('int64')

df.head()
    
#df['Peak'].astype(int)
#df['All Time Peak'].astype(int)
```

```python
df.info()
```

```python
#Organizando descendentemente los promedios

df.sort_values(by='Average gross', ascending=False).head()
```

```python
df.columns
```

```python
df_backup = df.copy()
newcols = []

for names in df_backup.columns:
    newcols.append(names.replace('\xa0', ' '))
    
df_backup.columns = newcols
df_backup.columns[3]
```

```python
grp_df_overview = df_backup.groupby(['Artist', 'Shows']).sum()['Actual gross']
grp_df_overview
```

```python
grp_df_NoShows = df_backup.groupby(['Artist']).sum()['Shows']
grp_df_NoShows
```

```python
grp_df_AmntRevs = df_backup.groupby(['Artist']).sum()['Actual gross']
grp_df_AmntRevs
```

```python
avg_per_artist = grp_df_AmntRevs.values / grp_df_NoShows.values

avg_per_artist
```

```python
df_revs_aggr = pd.DataFrame(
    {'Shows': grp_df_NoShows,
     'Revenue' : grp_df_AmntRevs,
     'Avg_Revenues': avg_per_artist.astype(int) 
    }, index = grp_df_NoShows.index   
)
df_revs_aggr.sort_values('Avg_Revenues', ascending= False, inplace=True)
df_revs_aggr
```
