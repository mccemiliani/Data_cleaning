```python
## Peak

### Reemplazar los NaN por 0
### Eliminar los datos con []
```

```python
df['Peak'].isna().sum()
```

```python
#Reemplazando con localizador
df.loc[df['Peak'] == 'NaN', 'Peak'] = 0
df['Peak']

```

```python
#Rellenando na para reemplazar

df['Peak'].fillna(0, inplace=True)
df['Peak']
```

```python
#Reemplazando con replace
df['Peak'] = df['Peak'].replace(to_replace=[None], value=0)
df['Peak']
```

```python
#Atacando los valores con llaves

def extraer_llaves(texto):
    
    if isinstance(texto, str):
        match = re.match(r'^(\d+)', texto)
        if match:
            return match.group(1)
    return texto

df['Peak'] = df['Peak'].apply(extraer_llaves)
df['Peak']
```

```python
df.head()
```

```python
df['All Time Peak'] = df['All Time Peak'].apply(extraer_llaves)
df['All Time Peak'].fillna(0, inplace=True)
df.head()
```

```python
#Cambiando formato a numero de Actual Gross
#df['Actual\xa0gross'] = df['Actual\xa0gross'].astype(int)

#Quitando simbolo y ,
df['Actual\xa0gross'] = df['Actual\xa0gross'].str.replace("$", "").str.replace(",", "")
df['Actual\xa0gross'] = df['Actual\xa0gross'].apply(extraer_llaves)
df['Actual\xa0gross'] = df['Actual\xa0gross'].astype('int64')
df['Actual\xa0gross']
```

```python
df.info()
```

```python
#Adjusted gross (in 2022 dollars)
#Retirando $, (,) y convirtiendo en formato número
#Convirtiendp Actual gross en dolares

df['Adjusted\xa0gross (in 2022 dollars)'] = df['Adjusted\xa0gross (in 2022 dollars)'].str.replace('$', '').str.replace(',', '')
df['Adjusted\xa0gross (in 2022 dollars)'] = df['Adjusted\xa0gross (in 2022 dollars)'].apply(extraer_llaves)
df['Adjusted\xa0gross (in 2022 dollars)'] = df['Adjusted\xa0gross (in 2022 dollars)'].astype('int64')


```
